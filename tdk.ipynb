```python
import numpy as np
def tok(txt):
    
    txt=txt.lower()

    for i in txt.split(' '):
        if i in to_be_removed:
            txt=txt.replace(i,"")

    txt=txt.replace('  ',', ')

    result=txt.split(', ')

    remove_punct_dict = dict((ord(punct), None) for punct in string.punctuation)

    #clear=[word for word in txt.split(' ') if not word in to_be_removed]



    result=[hunStemmer.stem(word.strip().translate(remove_punct_dict)).strip() for word in result if word!='']

    return(result)
```

```python
def azonos(str1,str2):
    
    gyakoriak=["magas","mgs","vér","kóros","krónikus","idült","betegség","általános","megbeteg","megbetegedés","rendellenesség","zavar","zav","dagan","daganat","műt","kór","tágul","tumor"]
    #for i in gyakoriak:
    #str1=str1.replace("betegség","")
    #str2=str2.replace("betegség","")
    
    cars1=str1.split(' ')
    cars2=str2.split(' ')
    if(Levenshtein.distance(str1,str2)<0.3*len(str1) and len(str1)>2):
        return True
    
    for i in cars1:
        for j in cars2 :
            if (Levenshtein.distance(i,j)<0.3*len(i) or (j.find(i)==0 or i.find(j)==0)) and len(i)>2 and not j in gyakoriak and not i in gyakoriak:
                return True
            
    
    
    return False
```
